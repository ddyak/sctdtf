```python
def plotter(decay, param, is_pull=False):
    fig=plt.figure(figsize=(20, 50))
    columns = 2
    rows = 6#len(energies)
    i = 1

    prefix = 'pull' if is_pull is True else 'fit' 
    energies = [0, 250, 1000]
    for energy in energies:
        img = mpimg.imread('fig/{}/{}/{}/{}_{}.png'.format(decay, methods[0], energy, prefix, param))
        fig.add_subplot(rows, columns, i);
        i += 1
        plt.imshow(img)
        img = mpimg.imread('fig/{}/{}/{}/{}_{}.png'.format(decay, methods[1], energy, prefix, param))
        fig.add_subplot(rows, columns, i);
        i += 1
        plt.imshow(img)
    plt.show()
```

```python
plotter('d_meson', 'd0_dpe')#, is_pull=True)
# plotter('kaon', 'ks_E')#, is_pull=True)
```
